# Scraping Abercrombie & Fitch

Little web scraper for the Abercrombie & Fitch website. The goal is to complete our database by adding some clothes in the unballanced categories.

```python
from bs4 import BeautifulSoup
import requests
import urllib.request
import sys
import os
import traceback
import re
import time
import json
import unidecode
```

We use our usual headers.

```python
headers = {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:62.0) Gecko/20100101 Firefox/62.0', 'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',}
```

```python
url = 'https://www.zara.com/fr/fr/homme-jeans-l659.html?v1=1079311'
html = requests.get(url, headers = headers)
```

Create a Beautiful Soup object.

```python
soup = BeautifulSoup(html.text, 'lxml')
```

We're just gonna get all the images on the url, as we don't want to scrap the whole website changing manually the url is not a big deal.

```python
links = soup.findAll("a", {'class':'item'})
urls = []
for ind, link in enumerate(links):
    lien = link['href']
    html_product = requests.get(lien, headers=headers)
    soupProduct = BeautifulSoup(html_product.text, 'lxml')
    img = soupProduct.find('a', {'class':'_seoImg main-image'})
    try:
        print("Downloading image no", ind)
        name = "https:" + img['href']
        print(name)
        urls.append(name)
    except:
        print("Error while downloading.")
    
    

```

Clean the list of urls: add 'https:' to links, delete the non-working links.

```python
for i in range(len(urls)):
    if not re.search('https:', urls[i]):
        print("Appending...")
        urls[i] = 'https:' + urls[i]
```

Now we can download all the images. Since we have only a few images, we order them by hand this time.

```python
for ind, web in enumerate(urls):
    file_name = '/media/arthur/DATA/Databases/Complement/Jeans/' + web.split("/")[-1]
    f = open(file_name, 'wb') 
    f.write(requests.get(web, allow_redirects=True).content)
    f.close()
    print("Downloading image no", ind)
        
```
